Now we are going to visualize the data set on a cartopy map. The red dots on the map will each represent a port.

Get the bounds of the map

```python
World_Port_Index.total_bounds
```

```python
plt.figure(figsize = (20, 12)) # Define frame size 
m = plt.axes(projection=ccrs.PlateCarree()) # Create the frame 


m.add_feature(cartopy.feature.LAND.with_scale('10m'), facecolor = "white")# Add color to map
m.add_feature(cfeature.OCEAN.with_scale('10m')) # Add the sea 
m.add_feature(cfeature.LAKES.with_scale('10m'), edgecolor = 'white') # Add lakes
m.add_feature(cfeature.RIVERS.with_scale('10m')) # Add rivers

# Add the World Port Index data set
m.scatter(World_Port_Index['LONGITUDE'], World_Port_Index['LATITUDE'],color="red", s = 10)

```

Get the bounds of the first 1000 polygones in the dataset.  We are going to use the bounds too se the map extent when plotting the data on the map.

```python
Exact_Earth_M1[:1000].total_bounds
```

Print the geometry polygone review the data.

```python
Exact_Earth_M1[:1]['geometry']
```

Plot the polygones on the map

```python
extent =[-11.81, 14. , 50.99,  61.98] # (left,right,down,up)

plt.figure(figsize = (20, 12)) # Define frame size 
m = plt.axes(projection=ccrs.PlateCarree()) # Create the frame 

m.set_extent(extent) # Add the extent, we set above, to the map



m.add_feature(cartopy.feature.LAND.with_scale('10m'), facecolor = "white")# Add color to map
m.add_feature(cfeature.OCEAN.with_scale('10m')) # Add the sea 
m.add_feature(cfeature.LAKES.with_scale('10m'), edgecolor = 'white') # Add lakes
m.add_feature(cfeature.RIVERS.with_scale('10m')) # Add rivers

# Add the World Port Index data set
m.add_geometries(Exact_Earth_M1[:3000]['geometry'], crs=ccrs.PlateCarree(), facecolor='g', edgecolor='red', alpha=0.8)
```

The plot above shows the first 3000 ships in the data set.

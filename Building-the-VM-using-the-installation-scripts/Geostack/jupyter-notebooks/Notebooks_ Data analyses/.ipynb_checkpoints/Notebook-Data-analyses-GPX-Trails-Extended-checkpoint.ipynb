```python
import gpxpy
import datetime
import numpy as np
import pandas as pd

from geopy import distance
from math import sqrt, floor
import haversine

import cartopy        
import cartopy.crs as ccrs 
import cartopy.feature as cfeature
import matplotlib.pyplot as plt

```

## Define generic functions

### Function: create_df
this function creates a data frame from the GPX files

parameters:
    - Data to convert to data frame

```python
def create_df(data):

    df = pd.DataFrame(columns=['lon', 'lat', 'alt', 'time'])

    for point in data:
        df = df.append({'lon': point.longitude,
                        'lat' : point.latitude, 
                        'alt' : point.elevation,
                        'time' : point.time}, ignore_index=True)
    return df
```

### Function: calculate_diffs
this function calculates the altitude difference, time difference and distance difference between all the data points.

parameters:
    - Data to calculate differences with
    - Dataframe to append results to

```python
def calculate_diffs(data,df):
    alt_dif = [0]
    time_dif = [0]
    dist_vin = [0]
    dist_hav = [0]
    dist_vin_no_alt = [0]
    dist_hav_no_alt = [0]
    dist_dif_hav_2d = [0]
    dist_dif_vin_2d = [0]

    for index in range(len(data)):
        if index == 0:
            pass
        else:
            start = data[index-1]

            stop = data[index]

            distance_vin_2d = distance.geodesic((start.latitude, start.longitude), (stop.latitude, stop.longitude)).m
            dist_dif_vin_2d.append(distance_vin_2d)

            distance_hav_2d = haversine.haversine((start.latitude, start.longitude), (stop.latitude, stop.longitude))*1000
            dist_dif_hav_2d.append(distance_hav_2d)

            dist_vin_no_alt.append(dist_vin_no_alt[-1] + distance_vin_2d)

            dist_hav_no_alt.append(dist_hav_no_alt[-1] + distance_hav_2d)

            alt_d = start.elevation - stop.elevation

            alt_dif.append(alt_d)

            distance_vin_3d = sqrt(distance_vin_2d**2 + (alt_d)**2)

            distance_hav_3d = sqrt(distance_hav_2d**2 + (alt_d)**2)

            time_delta = (stop.time - start.time).total_seconds()

            time_dif.append(time_delta)

            dist_vin.append(dist_vin[-1] + distance_vin_3d)

            dist_hav.append(dist_hav[-1] + distance_hav_3d)
            
            
    # Append results to dataframe         
    df['dis_vin_2d'] = dist_vin_no_alt 
    df['dist_hav_2d'] = dist_hav_no_alt
    df['dis_vin_3d'] = dist_vin
    df['dis_hav_3d'] = dist_hav
    df['alt_dif'] = alt_dif
    df['time_dif'] = time_dif
    df['dis_dif_hav_2d'] = dist_dif_hav_2d
    df['dis_dif_vin_2d'] = dist_dif_vin_2d
    
    #Print the results
    print('Vincenty 2D : ', dist_vin_no_alt[-1])
    print('Haversine 2D : ', dist_hav_no_alt[-1])
    print('Vincenty 3D : ', dist_vin[-1])
    print('Haversine 3D : ', dist_hav[-1])
    print('Total Time : ', floor(sum(time_dif)/60),' min ', int(sum(time_dif)%60),' sec ')
```

### Function: calculate_avg_speed
this function calculates the altitude difference, time difference and distance difference between all the data points.

parameters:
    - df =Dataframe from which you want the avg speed

```python
def calculate_avg_speed(df):
    
    df['dist_dif_per_sec'] = df['dis_dif_hav_2d'] / df['time_dif']
    
    df['spd'] = (df['dis_dif_hav_2d'] / df['time_dif']) * 3.6
    
    df_with_timeout = df[df['dist_dif_per_sec'] > 0]
    
    avg_km_h = (sum((df_with_timeout['spd'] * 
                 df_with_timeout['time_dif'])) / 
            sum(df_with_timeout['time_dif']))
    
    df['avg_spd'] = avg_km_h
    
    return avg_km_h
```

### Function: clean_convert_to_json
this function cleans the transformed data frame to only have the columns we need. Then it exports the data frame to JSON which we need to import the data in our data store.

parameters:
    - df = Dataframe which you want to clean
    - name = Name of output JSON file

```python
def clean_convert_to_json(df,name):
    df_clean = df[['lon','lat','alt','time','spd','avg_spd']]
    output_path = "../Data/Trail_JSON/" + name
    df_clean.to_json(output_path,orient='records')
    
```

### Function: basic_analyses
This function performs a basic GPX data analyses on an input file and returns a populated dataframe

parameters:
    - input_file = File location of the GPX file

```python
def basic_analyses(input_file):
    file = open(input_file, 'r')
    parsed_file = gpxpy.parse(file)
    print('\n--------------------------------BASIC INFORMATION-------------------------------\n\n')
    print("File info: " + str(parsed_file.tracks)+'\n')
    print("Tracks in file:" + str(len(parsed_file.tracks))+'\n')
    print("Segments in track: " + str(len(parsed_file.tracks[0].segments))+'\n')
    print("DataPoints in track: " + str(len(parsed_file.tracks[0].segments[0].points))+'\n')
    
    data = parsed_file.tracks[0].segments[0].points
    print("Start position and date :" + str(data[0])+'\n')
    print("End position and date :"  + str(data[-1])+'\n')
    df = create_df(data)
    print('\n--------------------------------CALCULATED DATA---------------------------------\n\n')
    calculate_diffs(data,df)
    print('avarage_speed: ' + str(calculate_avg_speed(df)))
    print('\n------------------------------COLUMNS AND DATA TYPES----------------------------\n\n')
    print(df.dtypes)
    print('\n-----------------------------DATA FRAME (first 2 rows)---------------------------\n\n')
    return df
    
    
    
    
```

### Function: create_map
This function creates a map and visualizes the inserted dataframe on the map.

parameters:
    - df = Dataframe to 

```python
plt.figure(figsize = (20, 12)) # Define frame size 
m = plt.axes(projection=ccrs.PlateCarree()) # Create the frame and add the projection 
m.coastlines(resolution='10m') # Create high resolution for coastal line
m.add_feature(cartopy.feature.LAND.with_scale('10m'), edgecolor='black', facecolor = "white")# Add color to map
m.add_feature(cfeature.OCEAN.with_scale('10m')) # Add the sea 
m.add_feature(cfeature.LAKES.with_scale('10m'), edgecolor = 'black') # Add lakes
m.add_feature(cfeature.RIVERS.with_scale('10m')) # Add rivers
m.add_feature(cfeature.BORDERS.with_scale('10m')) # Add borders


m.scatter(Biesbosch_df['lon'],Biesbosch_df['lat'],color='red',s=1) # Biesbosch rood 

m.scatter(ZeelandMNV['lon'],ZeelandMNV['lat'],color='blue',s=1)

m.scatter(Camper_Zeeland['lon'],Camper_Zeeland['lat'],color='green',s=1)

m.scatter(Biesbosch_Hank['lon'],Biesbosch_Hank['lat'],color='orange',s=1)

m.scatter(BBMuseum['lon'],BBMuseum['lat'],color='purple',s=1)

m.scatter(Dikkenberg_Libellen['lon'],Dikkenberg_Libellen['lat'],color='',s=1)

m.scatter(Dikkenberg_Nacht['lon'],Dikkenberg_Nacht['lat'],color='',s=1)
```

```python
def create_map(df):
    plt.figure(figsize = (20, 12)) # Define frame size 
    m = plt.axes(projection=ccrs.PlateCarree()) # Create the frame and add the projection 

    m.coastlines(resolution='10m') # Create high resolution for coastal line
    m.add_feature(cartopy.feature.LAND.with_scale('10m'), edgecolor='black', facecolor = "white")# Add color to map
    m.add_feature(cfeature.OCEAN.with_scale('10m')) # Add the sea 
    m.add_feature(cfeature.LAKES.with_scale('10m'), edgecolor = 'black') # Add lakes
    m.add_feature(cfeature.RIVERS.with_scale('10m')) # Add rivers
    m.add_feature(cfeature.BORDERS.with_scale('10m')) # Add borders

    # Add the first Swedish Crane set
    m.scatter(df['lon'], df['lat'],color="red", s = 5)
    return m
```

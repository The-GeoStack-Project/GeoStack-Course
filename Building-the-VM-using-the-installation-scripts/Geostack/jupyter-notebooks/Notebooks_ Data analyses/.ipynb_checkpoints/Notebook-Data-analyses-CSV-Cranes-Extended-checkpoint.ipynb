## Define generic functions

### Function: compare_sets()
This function compares the columns of 2 datasets and returns "are equal" if the columns of both datasets are the same. <br> The function returns the differences if the columns of both sets are not equal.

- Parameter 1 = the first set you want to compare<br>
- Parameter 2 = the second set you want to compare

```python
def compare_sets(set1,set2):
    
    # Create an empty list to append differences to
    diffs_set2=["IN SWEDISH BUT NOT IN GERMAN:",'']
    diffs_set1 =["IN GERMAN BUT NOT IN SWEDISH:",'']
    
    # Create an empty list to append equalities to
    eqs = ["IN BOTH SETS:",'']
    
    # Check is sets are the same
    # If True return "sets are equal"
    if set1.equals(set2):
        return "Sets are equal"
        
    # Compare sets and append diffs and eqs to corresponding lists
    else:
        for x in set1:
            if x in set2:
                eqs.append(x)
            if x not in set2:
                diffs_set1.append(x)
        for x in set2:
            if x not in set1:
                diffs_set2.append(x)
    return diffs_set1, diffs_set2,eqs
```

###  Function to convert to GeoJSON using Pandas
This function is used to convert a pandas dataframe to GeoJSON.

- Parameter 1 = Dataframe to convert
- Parameter 2 = The properties column
- Parameter 3 = The latitude column
- Parameter 4 = The longitude column
- Parameter 5 = The altitude column

```python
def df_to_geojson(df, properties, lat, lon):
    
    # Assign a featureCollection object to variable "geojson"
    geojson = {'type':'FeatureCollection', 'features':[]}
    
    # Iterate (Loop)trough all the rows in the dataframe
    for _, row in df.iterrows():
        
        # For each row in the dataframe
        # Assign a feature object to variable feature.
        # The feature object contains a properties object
        # and a geometry object of type "Point"
        # The geometry object contains a coordinates object
        feature = {'type':'Feature',
                   'properties':{},
                   'geometry':{'type':'Point',
                               'coordinates':[]}}
        
        # For each row in the dataframe 
        # Assign the values of the longitude and latitude columns
        # To the geometry object
        feature['geometry']['coordinates'] = [row[lon],row[lat]]
        
        # For each row in the dataframe
        # For each collumn in the properties list
        for prop in properties:
            
            # Assign the value of the column to
            # the properties object 
            feature['properties'][prop] = row[prop]
            
        # Append all the properties objects to
        # the features object. 
        geojson['features'].append(feature)
        
    # Return the created GeoJSON
    return geojson
```

### Function: basic_analysis
This function performs a basic data analyses on a input dataframe. <br>
- Parameter 1 = The dataframe to perform the basic analysis on

```python
def basic_analysis(df):
    
    print('\n--------------------------------BASIC INFORMATION-------------------------------\n\n')
    
    print("Dataframe length: " + str(len(df))+'\n') 
    
    print("Dataframe columns:" + str(len(df.columns))+'\n')
    
    print("Start date of study :" + str(df[:1].timestamp)+'\n')
    
    print("Start position (latitude) :" + str(df[:1]['location-lat'])+'\n')
    
    print("Start position (longitude) :" + str(df[:1]['location-long'])+'\n')
    
    print("End date of study :"  + str(df[-1:].timestamp)+'\n')
    
    print("End position (latitude) :" + str(df[-1:]['location-lat'])+'\n')
    
    print("End position (longitude) :" + str(df[-1:]['location-long'])+'\n')
    
    print('\n------------------------------COLUMNS AND DATA TYPES----------------------------\n\n')
    
    print(df.dtypes)
    
    print('\n------------------------------FIRST ROW OF DATAFRAME----------------------------\n\n')
    
    return df[:1]
```

## Export Swedish Crane datasets to GeoJSON

```python
# Define the columns we want to add to
# the properties object in our GeoJSON
properties = ['event-id', 
              'study-name',
              'timestamp',
              'visible',
              'ground-speed',
              'heading',
              'height-above-ellipsoid',
              'individual-taxon-canonical-name',
              'sensor-type',
              'tag-voltage',              
              'individual-local-identifier']

# Run the function and assign the output to variable "geojson"
# Parameter 1 = The dataframe to convert
# Parameter 2 = A list of column names to add to the properties object
# Parameter 3 = The column name of the latitude 
# Parameter 4 = The column name of the longitude
geojson_Frida = df_to_geojson(Filtered_SW_Crane_Frida, properties,
                        'location-lat',
                        'location-long')
geojson_Agnetha = df_to_geojson(Filtered_SW_Crane_Agnetha, properties,
                        'location-lat',
                        'location-long')

geojson_Cajsa = df_to_geojson(Filtered_SW_Crane_Cajsa, properties,
                        'location-lat',
                        'location-long')

# Assign the location and filename we want to create to variable "file"
file_Frida ='../Data/Crane_GeoJSON/20181003_Dataset_SV_GPS_Crane_9381_STAW_Crane_RRW-BuGBk_Frida.json'
file_Agnetha = '../Data/Crane_GeoJSON/20181003_Dataset_SV_GPS_Crane_9472_STAW_Crane_RRW-BuGR_Cajsa.json'
file_Cajsa = '../Data/Crane_GeoJSON/20181003_Dataset_SV_GPS_Crane_9407_STAW_Crane_RRW-BuGBk_Agnetha.json'

# Write the output of the function to a file.
# The indent parameter makes sure
# the geojson is human readable (easy to read)
with open(file_Frida, 'w') as file:
    json.dump(geojson_Frida, file, indent=2) 
with open(file_Agnetha, 'w') as file:
    json.dump(geojson_Agnetha, file, indent=2) 
with open(file_Cajsa, 'w') as file:
    json.dump(geojson_Cajsa, file, indent=2) 
```

#### Info on sensors:
- Bar: Baromatric sensor used to measure the pressure on an object.
- Gps: Sensor used for data related to GPS
- Mag: A MEMS magnetic field sensor is a small-scale microelectromechanical systems (MEMS) device for detecting and measuring magnetic fields (Magnetometer). Many of these operate by detecting effects of the Lorentz force: a change in voltage or resonant frequency may be measured electronically, or a mechanical displacement may be measured optically. Compensation for temperature effects is necessary. Its use as a miniaturized compass may be one such simple example application.
- GLS: A light level geolocator, light-level logger or GLS (global location sensor) is essentially a lightweight, electronic archival tracking device, usually used in bird migration research to map migration routes, identify important staging areas, and sometimes provide additional ecological information. A geolocator periodically records ambient light level (solar irradiance) to determine location.
- Orn: Sensor used for tranmssion protocols.

```python
# Define the columns we want to add to
# the properties object in our GeoJSON
properties = ['event-id', 
              'study-name',
              'timestamp',
              'visible',
              'ground-speed',
              'heading',
              'height-above-msl',
              'individual-taxon-canonical-name',
              'sensor-type',
              'tag-voltage',              
              'individual-local-identifier']

# Run the function and assign the output to variable "geojson"
# Parameter 1 = The dataframe to convert
# Parameter 2 = A list of column names to add to the properties object
# Parameter 3 = The column name of the latitude 
# Parameter 4 = The column name of the longitude
geojson_181527 = df_to_geojson(Filtered_GE_Crane_181528, properties,
                        'location-lat',
                        'location-long')
geojson_181528= df_to_geojson(Filtered_GE_Crane_181527, properties,
                        'location-lat',
                        'location-long')

# Assign the location and filename we want to create to variable "file"
file_181527 ='../Data/Crane_GeoJSON/20180928_Dataset_DE_GPS_Crane_181527_iCora_Crane_13_BuBuBr-YBuBk.json'
file_181528 = '../Data/Crane_GeoJSON/20191103_Dataset_DE_GPS_Crane_181528_iCora_Crane_15_BuBuBr-WYW_Lotta.json'

# Write the output of the function to a file.
# The indent parameter makes sure
# the geojson is human readable (easy to read)
with open(file_181527, 'w') as file:
    json.dump(geojson_181527, file, indent=2) 
with open(file_181528, 'w') as file:
    json.dump(geojson_181528, file, indent=2) 
```

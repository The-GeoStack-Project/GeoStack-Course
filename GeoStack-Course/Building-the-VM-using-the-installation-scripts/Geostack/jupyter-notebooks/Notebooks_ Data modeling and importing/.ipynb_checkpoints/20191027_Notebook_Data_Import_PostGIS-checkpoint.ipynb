## Import By hand

---

```python
def fetch_data(query):
    conn = psycopg2.connect(host='localhost',
                        dbname='Shapefile_Database',
                        user='mod_wms',
                        password='mod_wms',
                        port=5432)
    cursor = conn.cursor()
    
    cursor.execute(query)
    return pd.DataFrame(cursor.fetchall())

    if(conn):
        cursor.close()
        conn.close()
        print("PostgreSQL connection is closed")
```

## Querying the data 
---

Select all ports in database.

```python
fetch_data('select * from WPI;')
```

Select all russian ports

```python
fetch_data("select * from WPI where country='RU';")
```

USE-CASE:
We want to display all ports belonging to:
- Russia = Brown
- China = Yellow 
- America = Blue
- Netherlands = Red

And al give them another color

```python
Russia = fetch_data("select latitude,longitude from WPI where country='RU';")
China = fetch_data("select latitude,longitude from WPI where country='CN';")
America = fetch_data("select latitude,longitude from WPI where country='US';")
Netherlands =  fetch_data("select latitude,longitude from WPI where country='NL';")
```

```python
import cartopy        
import cartopy.crs as ccrs 
import cartopy.feature as cfeature 
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker 
```

```python
plt.figure(figsize = (20, 20)) # Define frame size 
m = plt.axes(projection=ccrs.PlateCarree()) # Create the frame 

m.add_feature(cartopy.feature.LAND.with_scale('10m'), facecolor = "white")# Add color to map
m.add_feature(cfeature.OCEAN.with_scale('10m')) # Add the sea 
m.add_feature(cfeature.LAKES.with_scale('10m'), edgecolor = 'white') # Add lakes
m.add_feature(cfeature.RIVERS.with_scale('10m')) # Add rivers


#- Russia = Brown
#- China = Red
#- America = Blue
#- Netherlands = Orange



m.scatter(Russia[1], Russia[0],color="brown", s = 20)

m.scatter(China[1], China[0],color="red", s = 20)

m.scatter(America[1], America[0],color="blue", s = 20)

m.scatter(Netherlands[1], Netherlands[0],color="orange", s = 20)

```

```python
# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -
# 0) Import the required modules or functions from modules.
import pandas as pd
# pandas is used to generate DataFrames from extracted data
```

```python
# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -
# 1) Import and Read the extracted dataset (in this case the file format is CSV)
Swedish_Crane_Dataframe = pd.read_csv("20191012_Data_Sets/Swedish_Crane.csv")
```

```python
# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -
# 2) Show the dataset 
Swedish_Crane_Dataframe
```

```python
# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -
# 3) Filter the dataset on the selected columns
Filtered_Swedish_Crane_Dataframe = Swedish_Crane_Dataframe[["event-id",
                                                           "timestamp",
                                                           "location-long",
                                                           "location-lat",
                                                           "ground-speed",
                                                           "heading",
                                                           "height-above-ellipsoid",
                                                           "tag-local-identifier"]]
```

```python
# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -
# 4)Show the filtered dataframe
Filtered_Swedish_Crane_Dataframe
```

```python
# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -
# 4)Export the filtered dataframe to a CSV file called "FILTERED_Swedish_Crane.csv"
# The variable index = None makes sure no index is added at on each row
# The variable header=True makes sure the column names are added to the CSV file 
Filtered_Swedish_Crane_Dataframe.to_csv("20191012_Data_Sets/FILTERED_Swedish_Crane.csv",index=None,header=True)
```
